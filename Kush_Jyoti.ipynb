```python
import numpy as np
import pandas as pd
import sklearn
```

```python
from sklearn.datasets import load_boston
kj = load_boston()
```

```python
kj.keys()
```

```python
boston = pd.DataFrame(kj.data, columns=kj.feature_names)
boston.head()
```

```python
boston['kjp'] = kj.target
boston.head()
```

```python
boston.isnull()
```

```python
boston.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
J = boston.drop('kjp', axis=1)
K = boston['kjp']
J_train, J_test, K_train, K_test = train_test_split(J, K, test_size = 0.15, random_state=5)
print(J_train.shape)
print(J_test.shape)
print(K_train.shape)
print(K_test.shape)   
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lin_model = LinearRegression()
lin_model.fit(J_train, K_train)
```

```python
K_train_predict = lin_model.predict(J_train)
rmse = (np.sqrt(mean_squared_error(K_train, K_train_predict)))

print("The model Performance for training set")
print('RMSE is {}'.format(rmse))
print("\n")

K_test_predict = lin_model.predict(J_test)
rmse = (np.sqrt(mean_squared_error(K_test, K_test_predict)))

print("The modelo performance for testing set")
print('RMSE is {}'.format(rmse))
```
